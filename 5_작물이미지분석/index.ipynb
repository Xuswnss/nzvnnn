````python
import os
import py7zr
````

````python
os.makedirs('../data/crop_data', exist_ok=True)


for name in ['NODefect_images', 'Defect_images', 'Mask_images']:
        src = f'../data/{name}.7z'
        if os.path.exists(src):
            print(f'{name} exist')
            with py7zr.SevenZipFile(src, mode='r') as z:
                z.extractall(path='../data/crop_data')
                print(f'{name} success')
        else:
            
            print(f'{name} not found')
````

````python
import cv2
import matplotlib.pyplot as plt

ok_files = sorted(os.listdir('../data/crop_data/NODefect_images'))
print(ok_files)

# 첫 번째 이미지 로드
img_path = os.path.join('../data/crop_data/NODefect_images', ok_files[0])


ok_bgr = cv2.imread(img_path)          # OpenCV: BGR 순서로 읽음
ok_rgb = cv2.cvtColor(ok_bgr, cv2.COLOR_BGR2RGB)  # BGR → RGB 변환
ok_bgr = cv2.imread(img_path)
if ok_bgr is None:
      print(f'이미지 로드 실패: {img_path}')
else:
      ok_rgb = cv2.cvtColor(ok_bgr, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(ok_bgr)                     # BGR 그대로 → 색상 왜곡
plt.title('BGR (raw)')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(ok_rgb)                     # RGB 변환 후 → 정상 색상
plt.title('RGB (corrected)')
plt.axis('off')

plt.tight_layout()
plt.show()

print(f'파일 수: {len(ok_files)}')
print(f'이미지 크기: {ok_bgr.shape}')  # (height, width, channels)
````
